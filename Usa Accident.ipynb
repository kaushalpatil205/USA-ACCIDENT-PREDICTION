```python
# import label encoder
from sklearn import preprocessing
# label_encoder object knows how to understand word labels
label_encoder=preprocessing.LabelEncoder()
```

```python
#encodes label in columns 'species'
data['Start_Time']=label_encoder.fit_transform(data['Start_Time'])
data['End_Time']=label_encoder.fit_transform(data['End_Time'])
data['Description']=label_encoder.fit_transform(data['Description'])
data['Side']=label_encoder.fit_transform(data['Side'])
data['Zipcode']=label_encoder.fit_transform(data['Zipcode'])
data['City']=label_encoder.fit_transform(data['City'])
data['Timezone']=label_encoder.fit_transform(data['Timezone'])
data['Wind_Direction']=label_encoder.fit_transform(data['Wind_Direction'])
data['Weather_Condition']=label_encoder.fit_transform(data['Weather_Condition'])
data['Sunrise_Sunset']=label_encoder.fit_transform(data['Sunrise_Sunset'])
data['Civil_Twilight']=label_encoder.fit_transform(data['Civil_Twilight'])
data['Nautical_Twilight']=label_encoder.fit_transform(data['Nautical_Twilight'])
data['Astronomical_Twilight']=label_encoder.fit_transform(data['Astronomical_Twilight'])

data['Start_Time'].unique()
data['End_Time'].unique()
data['Description'].unique()
data['Side'].unique()
data['Zipcode'].unique()
data['City'].unique()
data['Timezone'].unique()
data['Wind_Direction'].unique()
data['Weather_Condition'].unique()
data['Sunrise_Sunset'].unique()
data['Civil_Twilight'].unique()
data['Nautical_Twilight'].unique()
data['Astronomical_Twilight'].unique()
```

```python
data.dtypes
```

```python
#convert float to integer
data['Start_Lat']=label_encoder.fit_transform(data['Start_Lat'])
data['Start_Lng']=label_encoder.fit_transform(data['Start_Lng'])
data['Distance(mi)']=label_encoder.fit_transform(data['Distance(mi)'])
data['Number']=label_encoder.fit_transform(data['Number'])
data['Temperature(F)']=label_encoder.fit_transform(data['Temperature(F)'])
data['Pressure(in)']=label_encoder.fit_transform(data['Pressure(in)'])
data['Wind_Speed(mph)']=label_encoder.fit_transform(data['Wind_Speed(mph)'])

data['Start_Lat'].unique()
data['Start_Lng'].unique()
data['Distance(mi)'].unique()
data['Number'].unique()
data['Temperature(F)'].unique()
data['Pressure(in)'].unique()
data['Wind_Speed(mph)'].unique()
```

```python
#convert 'playoffs' column to integer
data.Traffic_Signal= data.Traffic_Signal.replace({True: 1, False: 0})
data.Amenity=data.Amenity.replace({True: 1, False: 0})
data.No_Exit= data.No_Exit.replace({True: 1, False: 0})
data.Junction=data.Junction.replace({True: 1, False: 0})
```
